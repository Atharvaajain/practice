```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/csgo (1).csv')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.describe().T
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['bomb_planted'] = le.fit_transform(df['bomb_planted'])
df['map'] = le.fit_transform(df['map'])
df['round_winner'] = le.fit_transform(df['round_winner'])
```

```python
df.info()
```

```python
x = df.iloc[:,:-1]
y = df['round_winner']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
new = pd.DataFrame(x_train)
new
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
```

```python
lda.fit(x_train,y_train)
```

```python
lda.transform(x_train)
```

```python
lda.coef_
```

```python
lda_coefficient = np.exp(np.abs(lda.coef_))
lda_coefficient
```

```python
lda_coefficient = lda_coefficient.flatten()
```

```python
feature_name = x.columns
feature_name
```

```python
plt.figure(figsize=(10,6))
plt.bar(feature_name,lda_coefficient)
plt.xticks(rotation=90)
plt.show()
```

```python
df_feature_score = pd.DataFrame({'feature':feature_name,'score':lda_coefficient})
df_feature_score
```

```python
important = df_feature_score.nlargest(20,'score').index
important
```

```python
x_train = x_train[:,important]
```

```python
x_test = x_test[:,important]
```

```python
pd.DataFrame(x_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
```

```python
from sklearn.metrics import *
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
tree.fit(x_train,y_train)
```

```python
treepred = tree.predict(x_test)
treepred
```

```python
accuracy_score(y_test,treepred)
```

```python
rf.fit(x_train,y_train)
```

```python
rfpred = rf.predict(x_test)
rfpred
```

```python
accuracy_score(y_test,rfpred)
```
